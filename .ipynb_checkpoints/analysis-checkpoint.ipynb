## Importing Modules

```python
import pandas as pd
import numpy as np
import datetime as dt
from scipy.stats import shapiro, describe
from arch.unitroot import PhillipsPerron
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-pastel')
plt.rcParams.update({'font.family': 'Times New Roman'})
plt.rcParams.update({'axes.grid': True, 'axes.linewidth': 0.5, 'axes.edgecolor': 'black'})
```

## 1. Data

```python
df = pd.read_csv("data/dataset.csv", index_col = 0)
df.index = pd.to_datetime(df.index)
```

```python
# Normalise Data between 0 and 1
def normalise(series):

    return (series - series.min()) / (series.max() - series.min())
```

## 2. Statistical Tests

We first look at the statistics for our index variables

We conduct the following statistical tests:
* The Shapiro-Wilk test for normality
* The Phillips-Perron test for stationarity

### 2.1 Sample Statistics

```python
def describe_data(series):

    result = describe(series)

    result =  {
            'nobs': result.nobs,
            'min': result.minmax[0],
            'max': result.minmax[1],
            'mean': result.mean,
            'variance': result.variance,
            'skewness': result.skewness,
            'kurtosis': result.kurtosis,
            }

    return pd.DataFrame(data = result, index = series.columns)
```

```python
print(describe_data(df))
```

### 2.2 Shapiro-Wilk Test

```python
def multivariate_shapiro(series):

    statistic = []
    pvalue = []

    for col in series.columns:
        result = shapiro(series[col])
        statistic.append(result.statistic)
        pvalue.append(result.pvalue)

    return pd.DataFrame(data = {'statistic': statistic, 'pvalue': pvalue}, index = series.columns)
```

```python
control_variables = ['INFLATION', 'DXY', 'SP500', '10Y_BOND', 'Term Spread', 'UNEMPLOYMENT', 'EXPINF10YR']

print(multivariate_shapiro(df[control_variables]))
```

### 2.3 Phillips Perron Test

```python
def multivariate_pp(series):

    statistic = []
    pvalue = []

    for col in series.columns:
        result = PhillipsPerron(series[col])
        statistic.append(result.stat)
        pvalue.append(result.pvalue)

    print(f"Null Hypothesis: {result.null_hypothesis}")
    print(f"Alternative Hypothesis: {result.alternative_hypothesis}")

    return pd.DataFrame(data = {'statistic': statistic, 'pvalue': pvalue}, index = series.columns)

print(multivariate_pp(df[control_variables]))
```

```python
stationarity_test_df = multivariate_pp(df[df.columns.difference(control_variables)])

stationarity_test_df[stationarity_test_df.pvalue > 0.05]
```

```python
fig, ax = plt.subplots(figsize = (16, 1))
sns.heatmap(multivariate_pp(df[df.columns.difference(control_variables)]).T, ax=ax);
```

```python
diff_df = df.diff().dropna()
# diff_df = df
```

## 3. Correlation Analysis

```python
fig, ax = plt.subplots(figsize = (12, 10))
sns.heatmap(diff_df[df.columns.difference(control_variables)].corr(), cmap = 'viridis', ax=ax);
ax.set_title('Query Correlation Plot', size = 16);
```

### 3.2 VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# the independent variables set
X = diff_df[diff_df.columns.difference(control_variables)]

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
						for i in range(len(X.columns))]

fig, ax = plt.subplots(figsize = (16, 5))
ax.bar(x=vif_data['feature'], height = vif_data['VIF'])
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90);
# ax.set_ylim(0, 50);
ax.axhline(y = 10, color = 'red', ls=':');
ax.set_title('Variance Inflation Factor', size = 16);
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=len(X.columns))
pca.fit(X)
X_pca = pca.transform(X)
fig, ax = plt.subplots(ncols = 2)
ax[0].bar(x = np.arange(len(pca.explained_variance_ratio_[:30])), height = pca.explained_variance_ratio_[:30]);
ax[1].bar(x = np.arange(len(pca.explained_variance_[:30])), height = np.cumsum(pca.explained_variance_ratio_[:30]));

ax[0].set_title('Explained Variance Ratio', size = 10);
ax[1].set_title('Cumulative Explained Variance Ratio', size = 10);

pca_df = pd.concat([diff_df, pd.DataFrame(X_pca[:, :30], index = diff_df.index)], axis=1)
pca_df.drop(diff_df.columns.difference(control_variables), axis=1, inplace=True)
```


## 3. Regression

### 3.2 Conducting OLS

```python
from sklearn.linear_model import LassoCV, RidgeCV
from sklearn.model_selection import KFold
dependent_column = 'INFLATION'
Y = pca_df[dependent_column]
X = pca_df[pca_df.columns.difference([dependent_column])]
X.columns = X.columns.astype(str)

X_train = X[:-28]
X_test = X[-28:]

Y_train = Y[:-28]
Y_test = Y[-28:]

cv = KFold(n_splits = 5, shuffle = False)
model = RidgeCV(store_cv_values = True)
# model =LassoCV(cv=cv, alphas = [1])

results = model.fit(X=X_train, y=Y_train)

coeffs = pd.DataFrame(data = {'Coeff': results.coef_}, index=results.feature_names_in_)

# test_set_mse = np.mean(results.mse_path_, axis=0)
test_set_mse = np.mean(results.cv_values_, axis = 1)
```

```python
print(coeffs[coeffs.Coeff > 0])
```

```python
from sklearn.metrics import mean_squared_error
y_pred = results.predict(X_test)

print(mean_squared_error(y_pred, Y_test))

Y_test_levels = df['INFLATION'].shift(1)[-28:] + Y_test
y_pred_levels = df['INFLATION'].shift(1)[-28:] + y_pred
```

```python
fig, ax = plt.subplots(figsize = (16, 5))

ax.plot(y_pred_levels, color = 'black');
# ax.plot(y_pred_levels + 1.96*np.sqrt(np.mean(test_set_mse)), color = 'red', ls=':');
# ax.plot(y_pred_levels - 1.96*np.sqrt(np.mean(test_set_mse)), color = 'red', ls=':');
ax.fill_between(x = y_pred_levels.index, y1 = y_pred_levels + 1.645*np.sqrt(np.mean(test_set_mse)),
                y2 = y_pred_levels - 1.645*np.sqrt(np.mean(test_set_mse)), alpha = 0.12, color = 'red')

ax.fill_between(x = y_pred_levels.index, y1 = y_pred_levels + 1.96*np.sqrt(np.mean(test_set_mse)),
                y2 = y_pred_levels - 1.96*np.sqrt(np.mean(test_set_mse)), alpha = 0.1, color = 'red')

ax.fill_between(x = y_pred_levels.index, y1 = y_pred_levels + 2.32*np.sqrt(np.mean(test_set_mse)),
                y2 = y_pred_levels - 2.32*np.sqrt(np.mean(test_set_mse)), alpha = 0.08, color = 'red')

ax.plot(Y_test_levels, color = 'blue');
```
